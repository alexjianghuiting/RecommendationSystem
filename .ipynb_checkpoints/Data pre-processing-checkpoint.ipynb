```python
#加Name
names = ['user_id','item_id','rating','timestamp']
dataDf = pd.read_csv('u.data',sep='\t',names=names)
dataDf = dataDf.drop('timestamp',1)
```

```python
#test and train
userTrainIndex = userDf.sample(frac=0.9).index
#true or false inside []
dataDf_train = dataDf[dataDf.user_id.isin(userTrainIndex.tolist())]
#drop the index
dataDf_test = dataDf.drop(dataDf_train.index)
print (dataDf_train.shape)
print (dataDf_test.shape)
dataDf_train.head()
```

```python
dataDf_train.to_csv('DATA2/ratings_train.csv')
dataDf_test.to_csv('DATA2/ratings_test.csv')
dataDf.to_csv('DATA2/ratings.csv')
```
